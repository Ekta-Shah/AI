### Numerical Columns

```python
numerical_df = data.select_dtypes(include = np.number)
```

```python
numerical_df['LotFrontage'].fillna(numerical_df['LotFrontage'].mean(),inplace=True)
numerical_df['GarageYrBlt'].fillna(numerical_df['GarageYrBlt'].mean(),inplace=True)
numerical_df['MasVnrArea'].fillna(numerical_df['MasVnrArea'].mean(),inplace=True)
```

```python
numerical_df.isnull().sum().sum()
```

```python
sns.heatmap(numerical_df.corr())
```

```python
## Selecting 10 features 
n = 7
numcols = numerical_df.corr().nlargest(n,'SalePrice')['SalePrice'].index
numcols
```

```python
sns.heatmap(numerical_df[numcols].corr(),annot=True)
```

```python
sns.pairplot(data = numerical_df[numcols])
```

```python
numerical_df[numcols].describe()
```

### Categorical Columns

```python
categorical_df = data.select_dtypes(exclude=np.number)
```

```python
categorical_df.head()
```

```python
categorical_df.describe()
```

```python
categorical_df.isnull().sum()
```

```python
categorical_df = categorical_df.drop(["Alley","FireplaceQu","PoolQC","Fence","MiscFeature"],1)
```

```python
categorical_df.isnull().sum()
```

```python
categorical_df["MasVnrType"].fillna(categorical_df["MasVnrType"].mode()[0],inplace=True)
categorical_df["BsmtQual"].fillna(categorical_df["BsmtQual"].mode()[0],inplace=True)
categorical_df["BsmtCond"].fillna(categorical_df["BsmtCond"].mode()[0],inplace=True)
categorical_df["BsmtExposure"].fillna(categorical_df["BsmtExposure"].mode()[0],inplace=True)
categorical_df["BsmtFinType1"].fillna(categorical_df["BsmtFinType1"].mode()[0],inplace=True)
categorical_df["BsmtFinType2"].fillna(categorical_df["BsmtFinType2"].mode()[0],inplace=True)
categorical_df["Electrical"].fillna(categorical_df["Electrical"].mode()[0],inplace=True)
categorical_df["GarageType"].fillna(categorical_df["GarageType"].mode()[0],inplace=True)
categorical_df["GarageFinish"].fillna(categorical_df["GarageFinish"].mode()[0],inplace=True)
categorical_df["GarageQual"].fillna(categorical_df["GarageQual"].mode()[0],inplace=True)
categorical_df["GarageCond"].fillna(categorical_df["GarageCond"].mode()[0],inplace=True)
```

```python
categorical_df.isnull().sum()
```

```python
sns.countplot(x="OverallQual",data=numerical_df)
```

```python
sns.boxplot(x="OverallQual",y="SalePrice",data=numerical_df)
```

```python
finaldf = pd.concat([numerical_df,categorical_df],1)
```

```python
plt.scatter(x=finaldf['GrLivArea'],y=finaldf['SalePrice'])
```

```python
plt.scatter(x= finaldf['TotalBsmtSF'], y = finaldf['SalePrice'])
```

```python
plt.scatter(x= finaldf['OverallQual'], y = finaldf['SalePrice'])
```

```python
catcolumns = list(categorical_df)
```

```python
finaldf['SalePrice']=np.log(finaldf['SalePrice'])
```

```python
nr_rows = 13
nr_cols = 3
target='SalePrice'
fig, axs = plt.subplots(nr_rows, nr_cols, figsize=(nr_cols*4,nr_rows*3))

for r in range(0,nr_rows):
    for c in range(0,nr_cols):  
        i = r*nr_cols+c
        if i < len(catcolumns):
            sns.boxplot(x=catcolumns[i], y=target, data=finaldf, ax = axs[r][c])
    
plt.tight_layout()    
plt.show()   
```
