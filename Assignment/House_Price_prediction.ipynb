```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
data = pd.read_csv('house_price_prediction.csv')
```

```
missvalues = data.isnull().sum()
```

```
data = data.drop(['Id'],1)
```

```
data1 = data.drop(['Alley','PoolQC','Fence','MiscFeature','FireplaceQu'],1)
```

```
missvalues = data1.isnull().sum()
```

```
list_of_cols=[]
for i in range(0,len(data1.columns)):
  if missvalues[i]>0:
    list_of_cols.append(missvalues.index[i])
    print(missvalues.index[i],missvalues[i])
```

```
data1['LotFrontage'].fillna(data1['LotFrontage'].mean(),inplace=True)
data1['GarageYrBlt'].fillna(data1['GarageYrBlt'].mean(),inplace=True)
data1['MasVnrArea'].fillna(data1['MasVnrArea'].mean(),inplace=True)

data1["MasVnrType"].fillna(data1["MasVnrType"].mode()[0],inplace=True)
data1["BsmtQual"].fillna(data1["BsmtQual"].mode()[0],inplace=True)
data1["BsmtCond"].fillna(data1["BsmtCond"].mode()[0],inplace=True)
data1["BsmtExposure"].fillna(data1["BsmtExposure"].mode()[0],inplace=True)
data1["BsmtFinType1"].fillna(data1["BsmtFinType1"].mode()[0],inplace=True)
data1["BsmtFinType2"].fillna(data1["BsmtFinType2"].mode()[0],inplace=True)
data1["Electrical"].fillna(data1["Electrical"].mode()[0],inplace=True)
data1["GarageType"].fillna(data1["GarageType"].mode()[0],inplace=True)
data1["GarageFinish"].fillna(data1["GarageFinish"].mode()[0],inplace=True)
data1["GarageQual"].fillna(data1["GarageQual"].mode()[0],inplace=True)
data1["GarageCond"].fillna(data1["GarageCond"].mode()[0],inplace=True)
```

```
numerical_df= data1.select_dtypes(include=np.number)
numerical_df.head()
y = data1['SalePrice']
numerical_df.drop(['SalePrice'],1)
numerical_df.corr()
# from sklearn.preprocessing import StandardScaler
# st=StandardScaler()
# cols=numerical_df.columns
# numerical_df=st.fit_transform(numerical_df)
# numerical_df=pd.DataFrame(numerical_df,columns=cols)
```

```
numerical_df
```

```
categorical_df=data1.select_dtypes(exclude=np.number)
categorical_df.head()
categorical_df=pd.get_dummies(categorical_df)
```

```
categorical_df.head()
```

```
X=pd.concat([numerical_df,categorical_df],1)
```

```
from sklearn.model_selection import train_test_split as tts
```

```
x_train,x_test,y_train,y_test=tts(X,y,test_size=0.2,random_state=42)
```

```
from sklearn.linear_model import LinearRegression
lnr = LinearRegression()
lnr.fit(x_train,y_train)
ypred = lnr.predict(x_test)
from sklearn.metrics import mean_squared_error
print(np.sqrt(mean_squared_error(y_test,ypred)))
```

```
from sklearn.model_selection import cross_val_score
```

```
from sklearn.ensemble import RandomForestRegressor
rfc=RandomForestRegressor(max_depth=4,n_estimators=500)
rfc.fit(x_train,y_train)
ypred = rfc.predict(x_test)
ypred2=rfc.predict(x_train)
from sklearn.metrics import r2_score
print("Training Accuracy",r2_score(y_train,ypred2))
print("Testing Accuracy",r2_score(y_test,ypred))
rfc
```

```
#from sklearn.metrics import mean_squared_error
cross_val_score(rfc,x_train,y_train,cv=10,scoring ='r2').mean()
```

```
x_test1=x_test[0:10]
y_test1=y_test[0:10]
ypred1=rfc.predict(x_test1)
print(y_test1)
print("Predict",ypred1)
#x_test1.columns
```

```
from sklearn.ensemble import GradientBoostingRegressor
gbrt=GradientBoostingRegressor(n_estimators=100) 
gbrt.fit(x_train, y_train) 
y_pred=gbrt.predict(x_test) 
    
```
