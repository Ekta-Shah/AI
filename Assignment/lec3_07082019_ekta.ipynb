### Function call for N - Queens Problem

```python
def under_attack(col, queens):
   # print(col,queens)
    #print("col in queens",col in queens)
    return col in queens or any(abs(col - x) == len(queens)-i for i,x in enumerate(queens))
```

```python
def solve(n):
    solutions = [[]]
    for row in range(n):
        solutions = [solution+[i+1] for solution in solutions for i in range(n) if not under_attack(i+1, solution)]
    return solutions
```

```python
print("Possible solutions : ")
count = 0
for answer in solve(n): 
    count = count + 1
    print(count,":",list(enumerate(answer, start=1)))
```
